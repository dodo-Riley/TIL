# 군집분석
## iris data 활용

```python
from sklearn.cluster import KMeans
from sklearn import cluster
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
# 데이터 불러오기
iris = datasets.load_iris()
```

```python
# 데이터 확인
iris
```

```python
# 피쳐 및 타겟 지정
data = iris.data
target = iris.target

data.shape, target.shape
```

### k-means

```python
# 학습
model = KMeans(n_clusters=3) # 군집개수 3개로 k-means 모델 객체 생성
model.fit(data) # 학습
```

```python
# 군집 결과 확인
model.labels_
```

```python
# 실제 타겟값 확인
iris.target
```

```python
# 각 군집별 데이터 확인
labels = model.labels_
print(f'label_0 : {data[labels==0]}', '\n')
print(f'label_1 : {data[labels==1]}', '\n')
print(f'label_2 : {data[labels==2]}', '\n')
```

```python
# 산점도로 군집 결과 확인(x축은 3번째 컬럼, y축은 4번째 컬럼 사용)
ldata_0 = data[labels==0]
plt.scatter(ldata_0[:,2], ldata_0[:,3],c='r',alpha=0.3, s=100,marker='o')

ldata_1 = data[labels==1]
plt.scatter(ldata_1[:,2], ldata_1[:,3],c='k',alpha=0.3, s=100,marker='x')

ldata_2 = data[labels==2]
plt.scatter(ldata_2[:,2], ldata_2[:,3],c='b',alpha=0.3, s=100,marker='*')
```

```python
ldata_0 = data[labels==0]
plt.scatter(ldata_0[:,2], ldata_0[:,3],c='r',alpha=0.3, s=100,marker='o')

ldata_1 = data[labels==1]
plt.scatter(ldata_1[:,2], ldata_1[:,3],c='k',alpha=0.3, s=100,marker='x')

ldata_2 = data[labels==2]
plt.scatter(ldata_2[:,2], ldata_2[:,3],c='b',alpha=0.3, s=100,marker='*')

plt.xlabel(iris.feature_names[2], fontsize= 'large')
plt.ylabel(iris.feature_names[3], fontsize= 'large')
plt.show()
```

### 계층적 군집(병합적)

```python
model_agg = cluster.AgglomerativeClustering(n_clusters=3, linkage='ward') # 군집수=3, 와드연결법(오차제곱합 최소) 사용한 객체 생성
model_agg.fit(data) # 학습
```

```python
# 군집 결과 확인
model_agg.labels_
```

```python
# 3번째 컬럼과 4번째 컬럼을 이용해 산점도 확인
labels_agg = model_agg.labels_
x_index = 2
y_index = 3
print(data[:,x_index], '\n')
print(data[:,y_index], '\n')
```

```python
data_x = data[:,x_index]
data_y = data[:,y_index]
```

```python
x_label = iris.feature_names[x_index]
y_label = iris.feature_names[y_index]
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_x[labels_agg==0], data_y[labels_agg==0], c='r', alpha=0.2, s=50, marker='o')
plt.scatter(data_x[labels_agg==1], data_y[labels_agg==1], c='b', alpha=0.2, s=50, marker='^')
plt.scatter(data_x[labels_agg==2], data_y[labels_agg==2], c='g', alpha=0.2, s=50, marker='*')
plt.xlabel(x_label, fontsize=15)
plt.ylabel(y_label, fontsize=15)
plt.title('Agg_clustering_ward', fontsize=20)
```

```python
# 3D 산점도를 통해 3개 피쳐에 대해 시각화해보기
z_index=1
z_label = iris.feature_names[z_index]
data_z = data[:,y_index]
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_x[labels_agg==0], data_y[labels_agg==0], data_z[labels_agg==0],c='r', alpha=0.2, s=50, marker='o')
ax.scatter(data_x[labels_agg==1], data_y[labels_agg==1], data_z[labels_agg==1],c='b', alpha=0.2, s=50, marker='^')
ax.scatter(data_x[labels_agg==2], data_y[labels_agg==2], data_z[labels_agg==2],c='g', alpha=0.2, s=50, marker='*')
ax.set_xlabel(x_label, fontsize=15)
ax.set_ylabel(y_label, fontsize=15)
ax.set_zlabel(z_label, fontsize=15)
ax.set_title('3D_Agg_clustering_ward', fontsize=20)
plt.show()
```

```python
# 덴드로그램
from scipy.cluster.hierarchy import dendrogram, linkage

z = linkage(data, 'ward')
plt.figure(figsize=(20,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('iris')
plt.ylabel('distance')
dendrogram(z,
          leaf_rotation=90, # 잎사위 노드 데이터값 출력 각도
          leaf_font_size=5) # 잎사귀 노드 데이터값 출력 사이즈
plt.show()
```

## 학업성취로 데이터 활용

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import cluster
```

```python
# 데이터 불러오기
academy = pd.read_csv('e:/Git_private_dodo_Riley/laptop_desktop/Code_SJH/통계학머신러닝_양진욱/data/academy.csv', encoding='cp949')
academy.columns=['num','kor','mat','eng','sci','con'] # 컬럼명 변경
academy
```

```python
academy.drop('num', axis=1, inplace=True) # num 컬럼은 인덱스이상의 의미가 없으므로 삭제
academy.head()
```

```python
# 데이터 확인
academy.shape
```

```python
academy.info()
```

```python
academy.describe()
```

```python
academy_arr = np.array(academy) # 데이터프레임을 어레이로 변환
academy.columns, academy_arr, academy.shape
```

```python
# 학습(k=means 활용)
model = cluster.KMeans(n_clusters=5) # 군집 수 5개로 객체 생성
model.fit(academy_arr) # 학습
```

```python
# 군집 내 데이터 확인
labels = model.labels_
ldata_0 = academy_arr[labels==0]
print(ldata_0)
print()
print(ldata_0[:,2])
print(ldata_0[:,3])
```

```python
# 일부 컬럼을 활용해 산점도를 그리고 군집 확인
ldata_0 = academy_arr[labels==0]
plt.scatter(ldata_0[:,2], ldata_0[:,3], c='r', alpha=0.5, marker='o')
ldata_1 = academy_arr[labels==1]
plt.scatter(ldata_1[:,2], ldata_1[:,3], c='g', alpha=0.5, marker='o')
ldata_2 = academy_arr[labels==2]
plt.scatter(ldata_2[:,2], ldata_2[:,3], c='b', alpha=0.5, marker='o')
ldata_3 = academy_arr[labels==3]
plt.scatter(ldata_3[:,2], ldata_3[:,3], c='y', alpha=0.5, marker='o')
ldata_4 = academy_arr[labels==4]
plt.scatter(ldata_4[:,2], ldata_4[:,3], c='k', alpha=0.5, marker='o')
plt.xlabel(academy.columns[2], fontsize=15)
plt.ylabel(academy.columns[3], fontsize=15)
plt.show()
```

```python
# 덴드로그램
from scipy.cluster.hierarchy import dendrogram, linkage

z = linkage(academy_arr, 'ward')
plt.figure(figsize=(20,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('cluster_subjects')
plt.ylabel('distance')
dendrogram(z,
          leaf_rotation=90,
          leaf_font_size=5)
plt.show()
```
