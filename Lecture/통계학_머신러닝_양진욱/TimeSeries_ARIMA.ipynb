```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
ts = pd.read_csv('e:/Git_private_dodo_Riley/laptop_desktop/Code_SJH/통계학머신러닝_양진욱/data/kospi3.csv',
           header=0,index_col=0, squeeze=True).sort_index()
ts.plot()
```

```python
plot_acf(ts)
plot_pacf(ts)
plt.show()
```

```python
# 차분
diff_1 = ts.diff(periods=1).iloc[1:]
diff_1.plot()
```

```python
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
# 차분 한 번 더 진행
diff_1_1 = diff_1.diff(periods=1).iloc[1:]
diff_1_1.plot()
```

```python
plot_acf(diff_1_1)
plot_pacf(diff_1_1)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(ts, order = (1,2,1))
fitted_model = model.fit(trend='c', full_output=True, disp=1)
fitted_model.summary()
```

```python
model = ARIMA(ts, order = (0,2,1))
fitted_model = model.fit(trend='c', full_output=True, disp=1)
fitted_model.summary()
```

```python
model = ARIMA(ts, order = (0,2,1))
fitted_model = model.fit(trend='nc', full_output=True, disp=1)
fitted_model.summary()
```

```python
fitted_model.plot_predict()
```

```python
# 예측
fore = fitted_model.forecast(steps=1)
print(fore)
```

```python
fore_10 = fitted_model.forecast(steps=10)
print(fore_10)
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline

df = pd.read_csv('e:/Git_private_dodo_Riley/laptop_desktop/Code_SJH/통계학머신러닝_양진욱/data/multiTimeline.csv', skiprows=1)
```

```python
df.head()
```

```python
df.columns = ['month','diet','gym','finance']
df.head()
```

```python
df['month'] = pd.to_datetime(df.month)
df.head()
```

```python
df.set_index('month',inplace=True)
```

```python
df.info()
```

```python
df.plot(figsize=(20,10), linewidth=5,fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
# 12 사이즈 이동 평균
diet = df[['diet']]
diet.rolling(12).mean().plot(figsize=(20,10), linewidth=5,fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
gym = df[['gym']]
gym.rolling(12).mean().plot(figsize=(20,10), linewidth=5,fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
diet = pd.concat([diet.rolling(12).mean(),gym.rolling(12).mean()],axis=1)
diet.plot(figsize=(20,10), linewidth=5,fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
# 다이어트 곡선에 대한 차분 
diet = df[['diet']]
diet.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
# 컬럼 간 상관관계
df.corr()
```

```python
# 3개 속성에 대한 차분
df.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
# 차분에 대한 상관관계
df.diff().corr()
```
