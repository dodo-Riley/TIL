```python
# 라이브러리 호출
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# 시드값 고정
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# 데이터 불러오기
df = pd.read_csv('./dataset/sonar.csv', header=None)
df.head()
```

```python
# 데이터 확인
df.info()
```

```python
df[60].unique()
```

```python
# 데이터를 넘파이 배열로 변환 및 타겟 분리
dataset = df.values
X = dataset[:,:-1].astype(float) # 타입이 object라서 float로 바꿔줘야함
y_obj = dataset[:,-1]

X[:5], y_obj[:5]
```

```python
# 타겟 라벨인코딩
le = LabelEncoder()
y = le.fit_transform(y_obj)
y[:3]
```

```python
# 모델 구성
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일 및 실행
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, batch_size=5)
```

```python
# 결과 출력
print(f'Accuracy: {model.evaluate(X, y)[1]:.4f}')
```

# 과적합 방지를 위해 hold out 적용

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['accuracy'])

model.fit(X_train, y_train, epochs=150, batch_size=5)
```

```python
# 결과 확인(테스트 데이터에 대해)
print("Test Accuracy: %.4f" % (model.evaluate(X_test, y_test)[1]))
```

# 생성한 모델을 저장하고 불러오기

```python
from sklearn.model_selection import train_test_split
from keras.models import load_model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=150, batch_size=5)
model.save('model.h5')  # 모델을 현재 작업 디렉토리에 저장

del model  # 현재 메모리에 저장된 모델 삭제
model = load_model('model.h5') # 저장한 모델 불러오기

# 결과 확인
print(f'Test Accuracy: {model.evaluate(X_test, y_test)[1]:.4f}')  
```

# k-fold 교차검증 사용

```python
from sklearn.model_selection import StratifiedKFold

n_fold = 10 # 나누는 그룹 개수
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=42)
```

```python
accuracy = [] # 각 사용데이터당 결과를 담을 리스트 생성

# 모델 구성, 컴파일, 실행
for train,test in skf.split(X,y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train],y[train], epochs=100, batch_size=5, verbose=0) # 모든 진행상황 출력하면 지저분하니까 verbose=0 설정
    
    k_acc = f'{model.evaluate(X[test],y[test])[1]:.4f}'
    accuracy.append(k_acc)

# 결과 출력
print(f'Total Accuracy : {accuracy}')
```

```python
# 오차함수 변경 : mse > 이진교차엔트로피

accuracy = [] # 각 사용데이터당 결과를 담을 리스트 생성

# 모델 구성, 컴파일, 실행
for train,test in skf.split(X,y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train],y[train], epochs=100, batch_size=5, verbose=0) # 모든 진행상황 출력하면 지저분하니까 verbose=0 설정
    
    k_acc = f'{model.evaluate(X[test],y[test])[1]:.4f}'
    accuracy.append(k_acc)

# 결과 출력
print(f'Total Accuracy : {accuracy}')
```
