```python
# 라이브러리 호출
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# 시드값 고정
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
df = pd.read_csv('./dataset/housing.csv',
                 delim_whitespace=True, # 공백의 크기가 얼마든 상관없이 공백을 구분자로 인식
                 header=None)
df.head()
```

```python
df.info()
```

```python
# 넘파이 배열로 변환, 타겟 분리
dataset = df.values
X = dataset[:,:-1]
y = dataset[:,-1]
```

```python
# 학습/테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 모델 생성, 컴파일, 학습
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1)) # 분류가 아니기 때문에 활성화함수 없음, 나온값을 그대로 사용해야함

model.compile(loss='mean_squared_error', optimizer='adam') # 평가지표 없음, 마찬가지로 예측 문제이기 때문
model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 예측 및 실제값과 비교
y_pred = model.predict(X_test).flatten() # 결과를 1차원으로 바궈주기 위해 flatten()

for i in range(10):
    label = y_test[i]
    pred = y_pred[i]
    print(f'실제값 : {label}, 예측값 : {pred}')
```
