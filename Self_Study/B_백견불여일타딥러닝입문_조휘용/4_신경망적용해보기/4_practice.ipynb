# 스케일링 효과 비교

## MNIST 데이터 사용

```python
from tensorflow.keras.datasets.mnist import load_data
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

(x_train,y_train),(x_test,y_test) = load_data('mnist.npz')
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=42)

# 데이터 전처리, 스케일링 미진행
x_train = x_train.reshape(-1,784)
x_val = x_val.reshape(-1,784)
x_test = x_test.reshape(-1,784)
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)

model = Sequential()
model.add(Dense(64, input_dim = 784, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(x_train, y_train, epochs=30, batch_size=128, validation_data=(x_val,y_val))

model.evaluate(x_test,y_test)
```

```python
(x_train,y_train),(x_test,y_test) = load_data('mnist.npz')
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=42)
# 데이터 전처리, 스케일링 진행
x_train = x_train.reshape(-1,784)/255
x_val = x_val.reshape(-1,784)/255
x_test = x_test.reshape(-1,784)/255
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)

model = Sequential()
model.add(Dense(64, input_dim = 784, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(x_train, y_train, epochs=30, batch_size=128, validation_data=(x_val,y_val))

model.evaluate(x_test,y_test)
```

## boston 데이터 사용

```python
from tensorflow.keras.datasets.boston_housing import load_data
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

(x_train, y_train), (x_test, y_test) = load_data(path='boston_housing.npz',test_split=0.2, seed=42)

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train,test_size = 0.33,random_state=42)

model = Sequential()
model.add(Dense(64, activation = 'relu', input_dim = 13))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mae'])
model.fit(x_train, y_train,epochs = 300, validation_data = (x_val, y_val))

model.evaluate(x_test, y_test)
```

```python
(x_train, y_train), (x_test, y_test) = load_data(path='boston_housing.npz',test_split=0.2, seed=42)

mean = np.mean(x_train, axis = 0)
std = np.std(x_train, axis = 0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train,test_size = 0.33,random_state=42)

model = Sequential()
model.add(Dense(64, activation = 'relu', input_dim = 13))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mae'])
model.fit(x_train, y_train,epochs = 300, validation_data = (x_val, y_val))

model.evaluate(x_test, y_test)
```

- 모두 스케일링을 진행해준 경우가 지표가 개선됨을 확인 가능

# 은닉층의 깊이에 따른 과적합 여부 확인

## Fashion_MNIST 데이터 사용

```python
from tensorflow.keras.datasets.fashion_mnist import load_data
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# 데이터 불러오기
(x_train, y_train), (x_test, y_test) = load_data()

# 데이터 전처리
x_train = x_train / 255
x_test = x_test / 255
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=42)
```

### 매우 적은 은닉층

```python
small_model = Sequential()

small_model.add(Flatten(input_shape = (28,28))) # (28, 28) -> .(28 * 28)
small_model.add(Dense(10, activation = 'relu'))
small_model.add(Dense(10, activation = 'softmax'))

small_model.compile(optimizer='adam',loss = 'categorical_crossentropy',metrics=['acc']) 

small_model.fit(x_train, y_train, epochs = 30, batch_size = 128, validation_data = (x_val, y_val))

small_model.evaluate(x_test,y_test)
```

### 중간 정도의 은닉층

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=30, batch_size=128, validation_data=(x_val,y_val))

model.evaluate(x_test,y_test)
```

### 매우 큰 은닉층

```python
big_model = Sequential()
big_model.add(Flatten(input_shape=(28,28)))
big_model.add(Dense(2048, activation = 'relu'))
big_model.add(Dense(1024, activation = 'relu'))
big_model.add(Dense(512, activation = 'relu'))
big_model.add(Dense(256, activation = 'relu'))
big_model.add(Dense(128, activation = 'relu'))
big_model.add(Dense(10, activation = 'softmax'))

big_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
big_model.fit(x_train, y_train, epochs=30, batch_size=128, validation_data=(x_val,y_val))

big_model.evaluate(x_test,y_test)
```

- 노드와 층의 수가 아주 많아지면 오히려 loss가 증가함 > 과적합
