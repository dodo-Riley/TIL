# 보스턴 주택 가격 데이터셋 다운

```python
from tensorflow.keras.datasets.boston_housing import load_data

(x_train, y_train), (x_test, y_test) = load_data(path='boston_housing.npz', test_split=0.2, seed=42)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# 데이터 전처리 및 검증 데이터 분할

```python
import numpy as np
from sklearn.model_selection import train_test_split
# 스케일링
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_train = (x_train-mean)/std
x_test = (x_test-mean)/std # train 평균과 표준편차로 스케일링해야함

# 데이터 분할
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.33, random_state=42)
```

# 모델 구성

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(64, input_dim=13, activation='relu')) # 책에서는 input_shape 썼음
model.add(Dense(32, activation='relu'))
model.add(Dense(1)) # 분류문제가 아니므로 활성화함수 없이 값 그대로 출력
```

# 모델 컴파일 및 학습

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
history = model.fit(x_train, y_train, epochs=300, validation_data=(x_val,y_val))
```

# 모델 평가

```python
print(f'test loss : {model.evaluate(x_test, y_test)[0]:.4f},test mae : {model.evaluate(x_test, y_test)[1]:.4f}')
```

# K-fold CV 적용

```python
# 데이터 수가 너무 적으므로 k폴드 교차검증 사용
from sklearn.model_selection import KFold

k=5
kfold = KFold(n_splits=k, shuffle=True, random_state=42) # kfold 객체 생성
# 책에서는 shuffle 옵션이 빠져있는데, 업데이트 된건지 shuffle없이 시드 고정이 안됨
# 아예 시드고정 옵션을 빼버리거나, shuffle=True와 함께 옵션 지정할 것

def get_model(): # 위에서 사용한 것과 동일한 구성의 모델을 반환하는 함수 정의
    model = Sequential()
    model.add(Dense(64, input_dim=13, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    return model

loss_list = [] # 각 시도마다 loss를 담을 리스트 생성
mae_list = [] # 각 시도마다 mae를 담을 리스트 생성

for train_idx, val_idx in kfold.split(x_train):
    x_train_fold, x_val_fold = x_train[train_idx], x_train[val_idx]
    y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]

    model = get_model() # 매 반복마다 모델을 새로 생성 > 그렇지 않으면 모든 데이터를 그냥 학습하는 모델일 뿐
    
    model.fit(x_train_fold, y_train_fold, epochs=300, validation_data=(x_val_fold, y_val_fold))
    
    test_loss, test_mae = model.evaluate(x_test, y_test)
    
    loss_list.append(test_loss)
    mae_list.append(test_mae)

loss_fold = np.mean(loss_list)
mae_fold = np.mean(mae_list)

print(f'test loss : {loss_fold:.4f}, test mae : {mae_fold:.4f}')

# k=3일 때 결과는 test loss : 13.8780, test mae : 2.6510
# k=5일 때는 test loss : 13.4396, test mae : 2.6860
# 더 좋아져야하는데.... 결과가 더 안좋다.....
# 데이터가 너무너무너무 적어서?
```
