# 다층 퍼셉트론과 XOR 게이트 구현

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.losses import mse

# 시드값 고정
tf.random.set_seed(42) 

# 데이터 생성
data = np.array([[0,0],[1,0],[0,1],[1,1]])
label = np.array([[0],[1],[1],[0]])

# 모델 구성
model = Sequential()
model.add(Dense(32, input_shape=(2,), activation='relu')) # 출력 노드32, 입력값은 1차원, 활성화함수 렐루
model.add(Dense(1, activation='sigmoid')) # 0/1 이진분류이므로 출력 노드 1, 활성화함수 시그모이드 

# 모델 컴파일
model.compile(loss = mse, optimizer = RMSprop(), metrics=['acc'])

# 모델 학습
model.fit(data, label, epochs=100)
```

# 여러가지 활성화함수 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# 시그모이드
def sigmoid(x):
    return 1/(1+np.exp(-x))

# 하이퍼볼릭 탄젠트
def tanh(x):
    return list(map(lambda x:math.tanh(x),x))

# relu : 0이하는 0, 이상은 그대로
def relu(x):
    result = []
    for ele in x:
        if ele <= 0:
            result.append(0)
        else:
            result.append(ele)
    return result

x = np.linspace(-4,4,100) # -4부터 4까지 100개로 나눈 값을 x로 지정
result_sigmoid = sigmoid(x)
result_tanh = tanh(x)
result_relu = relu(x)
plt.subplots(constrained_layout=True)
plt.subplot(131), plt.plot(x, result_sigmoid)
plt.subplot(132), plt.plot(x, result_tanh)
plt.subplot(133), plt.plot(x, result_relu)
plt.show()
```

# 경사하강법 실험

```python
import numpy as np
import matplotlib.pyplot as plt

lr_list = [0.001, 0.1, 0.5, 0.9] # 학습률 리스트

def get_derivative(lr):
    w_old = 2 # 초기 가중치 = 2
    derivative = [w_old] # 가중치를 저장할 리스트
    y=[w_old**2] # 오차함수
    
    for i in range(1,10):
        dev_value = w_old*2 # 오차함수의 미분
        
        w_new = w_old - lr * dev_value # 학습률만큼 가중치 업데이트
        w_old = w_new # 가중치 대체
        
        derivative.append(w_old) # 가중치 리스트에 추가
        y.append(w_old**2) # 오차 리스트에 추가
    return derivative, y

x = np.linspace(-2,2,50)
x_square = [i**2 for i in x]

fig = plt.figure(figsize=(12,7))

for i, lr in enumerate(lr_list): # enumerate는 인덱스와 원소를 반환해줌
    derivative, y = get_derivative(lr)
    ax = fig.add_subplot(2,2,i+1)
    ax.scatter(derivative, y, color='r')
    ax.plot(x, x_square)
    ax.title.set_text('lr = '+str(lr))
plt.show()
    
```
