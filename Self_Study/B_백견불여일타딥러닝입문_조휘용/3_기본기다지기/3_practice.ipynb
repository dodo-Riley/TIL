# 단층 퍼셉트론으로 XOR 게이트 실습(결과값이 제대로 나오지 않는 것이 정상)

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 시드값 고정
tf.random.set_seed(42)

# 데이터 생성
x = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# 모델 구성
model = Sequential()
model.add(Dense(1, input_shape = (2, ), activation = 'linear'))

# 모델 컴파일
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])

# 모델 학습
model.fit(x, y, epochs = 100)
```

```python
# 결과 출력
print(f'loss : {model.evaluate(x,y)[0]}, accuracy : {model.evaluate(x,y)[1]}')
print(f'예측값 : {model.predict(x)}')
```

# 학습률에 따른 다층 퍼셉트론 XOR 게이트 문제 확인

```python
from tensorflow.keras.optimizers import Adam

# 시드값 고정
tf.random.set_seed(42)

# 데이터 생성
x = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# 모델 구성
model = Sequential()
model.add(Dense(32, input_shape = (2, ), activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

# 모델 컴파일
model.compile(optimizer = Adam(lr=0.1), loss = 'mean_squared_error', metrics = ['accuracy'])

# 모델 학습
model.fit(x, y, epochs = 100)

# 결과 출력
print(f'loss : {model.evaluate(x,y)[0]}, accuracy : {model.evaluate(x,y)[1]}')
print(f'예측값 : {model.predict(x)}')
```
