# 지도에 마커 표시하기

```python
df = pd.read_excel('e:/data_파이썬머신러닝판다스데이터분석/part4/서울지역 대학교 위치.xlsx')
df.set_index(df.iloc[:,0], inplace=True) # df의 첫째컬럼을 인덱스로 지정

seoul_map = folium.Map(location=[37.55,126.96], tiles='Stamen Terrain', zoom_start=12)

for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat,lng], # [위도,경도] 
                  popup=name # 마커를 클릭 시 출력 데이터,
                 ).add_to(seoul_map) #  seoul_map 에 추가
seoul_map
```

```python
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat,lng], # [위도,경도] 
                        radius=10, # circlemarker 반지름
                        color='brown', # 외곽선 색상
                        fill=True, # 내부 채우기 여부
                        fill_color='coral', # 내부 색상
                        fill_opacity=0.7, # 내부색상 투명도
                        popup=name # 마커를 클릭 시 출력 데이터,
                         ).add_to(seoul_map) #  seoul_map 에 추가
seoul_map
```

# 지도 영역에 단계구분도 표시하기

```python
file_path = ('e:/data_파이썬머신러닝판다스데이터분석/part4/경기도인구데이터.xlsx')
df = pd.read_excel(file_path, index_col='구분') # 구분 열을 인덱스로 불러오기
df.columns = df.columns.map(str) # 컬럼명 문자열로 변환

geo_path = ('e:/data_파이썬머신러닝판다스데이터분석/part4/경기도행정구역경계.json')

# json 파일 불러오기
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
    

# 지도 객체 생성    
g_map = folium.Map(location=[37.5502,126.982], tiles='Stamen Terrain', zoom_star=9)

# 년도 지정
year='2011'

# 단계구분도 그리기
folium.Choropleth(geo_data=geo_data, # 지형 데이터 
                 data = df, # 사용할 데이터
                 columns = [df.index, year], # 데이터 비교 기준 컬럼(사용할 데이터에서)과 사용할 컬럼
                 fill_color = 'YlOrRd', # 내부 색상
                 fill_opacity = 0.7, # 내부 투명도
                 line_opacity = 0.3, # 선 투명도
                 threshold_scale = [10000,100000,300000,500000,700000], # 색상 구분 기준
                 key_on = 'feature.properties.name', # 데이터 비교 기준(지형 데이터에서)
                 ).add_to(g_map)

g_map
```
