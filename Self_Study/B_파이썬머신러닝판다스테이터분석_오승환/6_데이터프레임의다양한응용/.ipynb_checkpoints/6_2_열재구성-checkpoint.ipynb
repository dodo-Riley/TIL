```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# 열 순서 변경

```python
# 데이터 불러오기
titanic = sns.load_dataset('titanic')
df = titanic.loc[0:4, 'survived':'age']
df
```

```python
# 컬럼명만 추출해 리스트로 변환
columns = list(df.columns) # 교재에는 df.columns.values를 했지만 굳이 values 필요없음
columns
```

```python
columns_sorted = sorted(columns) # columns 요소 정렬(알파벳 오름차순), sorted는 정렬된 결과를 반환할 뿐 원본은 그대로, sort() 메소드는 정렬 결과로 바꿈
df_sorted = df[columns_sorted] 
df_sorted
```

```python
columns_reversed = reversed(columns) # 기존 순서의 반대 순서로 정렬
df_reversed = df[columns_reversed]
df_reversed
```

```python
columns_customed = ['pclass', 'sex', 'age', 'survived'] # 원하는 순서대로 변경
df_customed = df[columns_customed]
df_customed
```

```python
# 참고 : 데이터프레임에 없는 열로 순서를 변경하거나 기존 컬럼 수와 다르게 지정하면 오류 발생
# 열 순서를 변경하고자 할때는 기존에 존재하는 컬럼이어야 한다
```

# 열 분리

```python
# 데이터 불러오기 및 확인
df = pd.read_excel('e:/data_파이썬머신러닝판다스데이터분석/part6/주가데이터.xlsx')
print(df.head(), '\n')
print(df.dtypes, '\n')
```

```python
df['연월일'] = df['연월일'].astype('str') # datetime 자료를 문자열로 변환
dates = df['연월일'].str.split('-') # 연월일 컬럼의 문자열 속성에 split을 적용, - 기호를 기준으로 문자열을 구분하고 리스트로 묶은 값을 가지는 시리즈로 반환
print(dates.head(), '\n')

# 참고 문자열 속성 대신 apply 이용하기
dates2 = df['연월일'].apply(lambda x:x.split('-'))
dates2.head()
```

```python
# 연월일 나눠서 새로운 컬럼 생성
df['연'] = dates.str.get(0) # 문자열 속성에 get() 적용, 인덱싱 번호로 해당 리스트의 값을 가져옴
df['월'] = dates.str.get(1)
df['일'] = dates.str.get(2)

# 참고1 : get() 사용하지 않고 인덱싱으로 가져오기
df['연2'] = dates.str[0]

# 참고2 : apply()메소드 활용
df['연3'] = dates.apply(lambda x:x[0])

df
```
