# 시리즈

```python
# 시리즈는 데이터프레임에서 한 컬럼만 있다고 생각할 수 있음
s = pd.Series(['amy',170,240])
s
```

```python
print(type(s))
print(s.index)
print(s.values)
s.index = ['name','height','footsize'] # 시리즈의 인덱스를 해당 리스트의 값으로 변경
print(s.index)
```

```python
s2 = pd.Series([10,20,30,40,50])
print('평균:',s2.mean())
print('최소값:',s2.min())
print('최대값:',s2.max())
print('중간값:',s2.median())
print('표준편차:',s2.std())
print(s2.describe())
```

```python
s3 = pd.Series([1,3,2,4,10])
s3 = s3.replace(10,5) # 시리즈 내 값 변경
s3
```

```python
s3.sort_values() # 값을 기준으로 오름차순 정렬
```

```python
s3.sort_values(ascending=False) # 값을 기준으로 내림차순 정렬
```

```python
s3.to_frame() # 시리즈를 컬럼이 하나인 데이터프레임으로 변경
```
