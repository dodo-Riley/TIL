# 클래스

```python
class Car: # 클래스 정의
    def __init__(self, b,m,c): # 생성자 : 객체가 생성될때 자동으로 호출되는 메소드
        self.brand = b
        self.model = m
        self.color = c
        
        print(self.brand, self.model, self.color, '출고')
    # 해당 클래스를 통해 생성된 객체에 수행될 수 있는 메소드 정의
    def turn_on(self):
        print(self.brand, '시동을 겁니다.')
        
    def turn_off(self):
        print(self.brand, '시동을 끕니다.')
    
    def drive(self):
        print(self.brand, '주행중입니다.')        
        
# 객체 생성
car1 = Car('현대자동차','소나타','화이트')
car2 = Car('르노삼성','SM7','블랙')

# 메소드 호출
car1.turn_on()
car1.turn_off()
car1.drive()

car2.turn_on()
car2.turn_off()
car2.drive()

# 객체에 사용가능한 메소드 조회
dir(Car)

# 특정 함수나 메소드의 사용법 조회
help(Car)
help(print)
a = [1,2,3]
help(a.append)
```

# 모듈과 패키지

```python
import pandas # pandas 모듈 호출
print(pandas.Series([1,2,3]))
print(pandas.DataFrame([1,2,3]))

import pandas as pd # pandas 모듈 호출하고 pd라는 이름으로 사용
print(pd.Series([1,2,3]))
print(pd.DataFrame([1,2,3]))

from pandas import Series, DataFrame # pandas 모듈 내 특정 함수를 직접 호출
print(Series([1,2,3]))
print(DataFrame([1,2,3]))

from pandas import * # 모듈 내 모든 함수 호출
print(Series([1,2,3]))
print(DataFrame([1,2,3]))
print(pd.Series([1,2,3]))
```

# 파이썬 내장 모듈

```python
import math
print(math.ceil(1.4)) # 정수로 올림
print(math.floor(1.7)) # 정수로 내림
print(math.sqrt(4)) # 제곱근
print(math.factorial(3)) # 팩토리얼
print(math.pi) # 원주율값

import random
print(random.randint(1,10)) # 1이상 10이하 중 랜덤한 정수 하나 반환
print(random.randrange(1,10)) # 1이상 10미만 중 랜덤정수 반환
print(random.randrange(10)) # 0이상 10미만 중 랜덤정수 반환
print(random.random()) # 0과 1사이 랜덤 실수 반환
print(random.uniform(1,10)) # 1이상 10미만 랜덤 실수 반환
print(random.choice([1,2,3])) # 해당 값 중 랜덤하게 한 가지 반환
print(random.choice('python')) # p,y,t,h,o,n 중 하나 반환
print(random.choice(range(1,101))) # 1이상 100이하 정수 중 하나 반환
print(random.sample([1,2,3],2)) # 해당 리스트 중 랜덤하게 2개 반환
print(random.sample('python',2)) # p,y,t,h,o,n 중 2개 반환
print(random.sample(range(1,101),2)) # 1이상 100이하 정수 중 2개 반환
a = [1,2,3,4,5]
random.shuffle(a) # 기존 순서를 랜덤하게 바꿈
print(a) 

import datetime
now = datetime.datetime.now() # 현재 날짜/시간 
print(now.year,'년')
print(now.month,'월')
print(now.day,'일')
print(now.hour,'시')
print(now.minute,'분')
print(now.second,'초')
print(now.strftime('%Y.%m.%d %H:%M:%S')) # 날짜 표현 형식 변경
print(now + datetime.timedelta(weeks=1, days=1, hours=1, minutes=1, seconds=1)) # 특정 시간이 지난 후 날짜 구하기
print(now + datetime.timedelta(days=100)) # 100일후
print(now + datetime.timedelta(days=-100)) # 100일전, 이 때 년도 단위는 계산이 불가

import time
tm = time.localtime() # 현재 날짜와 시간이 time 객체로 생성
print(tm)
print(time.localtime().tm_year)
print(time.localtime().tm_mon)
print(time.localtime().tm_mday)
print(time.localtime().tm_hour)
print(time.localtime().tm_min)
print(time.ctime())  # 현재 날짜와 시간 
print(3)
time.sleep(1) # 1초간 대기
print(2)
time.sleep(1)
print(1)
time.sleep(1)
print('srart')

# 내장 함수 확인하기
# dir(__builtins__)

# 외부 모듈 설치하기 및 확인
# pip install 모듈명
# pip list
```
